```python
# =====================================================
# STEP 4 : DATA CLEANING
# =====================================================

print("="*60)
print("DATA CLEANING")
print("="*60)

# -----------------------------------------------------
# Create Copy of Dataset
# -----------------------------------------------------

df_clean = df.copy()

print("\nOriginal Shape :", df.shape)

# -----------------------------------------------------
# 1. Check Missing Values
# -----------------------------------------------------

print("\nMissing Values Before Cleaning")

display(df_clean.isnull().sum())

print(
    "\nTotal Missing Values :",
    df_clean.isnull().sum().sum()
)

# -----------------------------------------------------
# 2. Handle Missing Values
# -----------------------------------------------------

# Numerical Columns

num_cols = df_clean.select_dtypes(
    include=['int64','float64']
).columns

for col in num_cols:

    if df_clean[col].isnull().sum() > 0:

        df_clean[col].fillna(
            df_clean[col].median(),
            inplace=True
        )

# Categorical Columns

cat_cols = df_clean.select_dtypes(
    include=['object']
).columns

for col in cat_cols:

    if df_clean[col].isnull().sum() > 0:

        df_clean[col].fillna(
            df_clean[col].mode()[0],
            inplace=True
        )

# -----------------------------------------------------
# 3. Verify Missing Values
# -----------------------------------------------------

print("\nMissing Values After Treatment")

display(df_clean.isnull().sum())

print(
    "\nTotal Missing Values :",
    df_clean.isnull().sum().sum()
)

# -----------------------------------------------------
# 4. Check Duplicate Records
# -----------------------------------------------------

duplicates = df_clean.duplicated().sum()

print(
    "\nDuplicate Records Before Removal :",
    duplicates
)

# -----------------------------------------------------
# 5. Remove Duplicate Records
# -----------------------------------------------------

df_clean.drop_duplicates(
    inplace=True
)

print(
    "\nShape After Duplicate Removal :",
    df_clean.shape
)

# -----------------------------------------------------
# 6. Standardize Text Data
# -----------------------------------------------------

cat_cols = df_clean.select_dtypes(
    include=['object']
).columns

for col in cat_cols:

    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.lower()
        .str.strip()
    )

print(
    "\nText Standardization Completed"
)

# -----------------------------------------------------
# 7. Verify Unique Values
# -----------------------------------------------------

print("\nUpdated Categories")

for col in cat_cols:

    print(f"\n{col}")

    print(df_clean[col].unique())

# -----------------------------------------------------
# 8. Check Data Types
# -----------------------------------------------------

print("\nData Types")

display(df_clean.dtypes)

# -----------------------------------------------------
# 9. Outlier Check (IQR Method)
# -----------------------------------------------------

print("\nOutlier Summary")

for col in num_cols:

    Q1 = df_clean[col].quantile(0.25)

    Q3 = df_clean[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    outliers = df_clean[
        (df_clean[col] < lower) |
        (df_clean[col] > upper)
    ].shape[0]

    print(
        f"{col} : {outliers} outliers"
    )

# -----------------------------------------------------
# 10. Final Dataset Shape
# -----------------------------------------------------

print("\nFinal Dataset Shape")

print(df_clean.shape)

# -----------------------------------------------------
# 11. Dataset Preview
# -----------------------------------------------------

print("\nCleaned Dataset Preview")

display(df_clean.head())

# -----------------------------------------------------
# Cleaning Summary
# -----------------------------------------------------

print("\nDATA CLEANING SUMMARY")

print("-"*50)

print(
    "Rows :", df_clean.shape[0]
)

print(
    "Columns :",
    df_clean.shape[1]
)

print(
    "Missing Values :",
    df_clean.isnull().sum().sum()
)

print(
    "Duplicate Records :",
    df_clean.duplicated().sum()
)

print("\nData Cleaning Completed Successfully")
```

```python
# =====================================================
# STEP 5 : ENCODE BINARY YES/NO COLUMNS
# =====================================================

print("="*60)
print("ENCODE BINARY YES/NO COLUMNS")
print("="*60)

# -----------------------------------------------------
# Create Copy
# -----------------------------------------------------

df_encoded = df_clean.copy()

# -----------------------------------------------------
# Binary Columns
# -----------------------------------------------------

binary_cols = [
    'mainroad',
    'guestroom',
    'basement',
    'hotwaterheating',
    'airconditioning',
    'prefarea'
]

print("\nBinary Columns")

print(binary_cols)

# -----------------------------------------------------
# Check Original Values
# -----------------------------------------------------

print("\nOriginal Value Counts")

for col in binary_cols:

    print(f"\nColumn : {col}")

    print(df_encoded[col].value_counts())

# -----------------------------------------------------
# Encode Yes / No
# -----------------------------------------------------

for col in binary_cols:

    df_encoded[col] = df_encoded[col].map({
        'yes': 1,
        'no': 0
    })

# -----------------------------------------------------
# Verify Encoding
# -----------------------------------------------------

print("\nEncoded Values")

display(df_encoded[binary_cols].head())

# -----------------------------------------------------
# Check Data Types
# -----------------------------------------------------

print("\nEncoded Column Data Types")

print(df_encoded[binary_cols].dtypes)

# -----------------------------------------------------
# Check Unique Values
# -----------------------------------------------------

print("\nUnique Values After Encoding")

for col in binary_cols:

    print(
        f"{col} :",
        df_encoded[col].unique()
    )

# -----------------------------------------------------
# Dataset Preview
# -----------------------------------------------------

print("\nDataset Preview")

display(df_encoded.head())

# -----------------------------------------------------
# Shape Check
# -----------------------------------------------------

print("\nDataset Shape")

print(df_encoded.shape)

# -----------------------------------------------------
# Summary
# -----------------------------------------------------

print("\nEncoding Completed Successfully")

print(
    "Encoded Columns :",
    len(binary_cols)
)
```

```python
# =====================================================
# STEP 6 : FEATURE ENGINEERING
# =====================================================

print("="*60)
print("FEATURE ENGINEERING")
print("="*60)

# -----------------------------------------------------
# Create Copy
# -----------------------------------------------------

df_fe = df_encoded.copy()

print("\nOriginal Shape :", df_fe.shape)

# -----------------------------------------------------
# 1. Total Rooms
# -----------------------------------------------------

df_fe['total_rooms'] = (
    df_fe['bedrooms'] +
    df_fe['bathrooms']
)

print("\nFeature Created : total_rooms")

# -----------------------------------------------------
# 2. Area Per Bedroom
# -----------------------------------------------------

df_fe['area_per_bedroom'] = (
    df_fe['area'] /
    df_fe['bedrooms']
)

print("Feature Created : area_per_bedroom")

# -----------------------------------------------------
# 3. Area Per Bathroom
# -----------------------------------------------------

df_fe['area_per_bathroom'] = (
    df_fe['area'] /
    df_fe['bathrooms']
)

print("Feature Created : area_per_bathroom")

# -----------------------------------------------------
# 4. Luxury Score
# -----------------------------------------------------

df_fe['luxury_score'] = (

    df_fe['guestroom'] +

    df_fe['basement'] +

    df_fe['hotwaterheating'] +

    df_fe['airconditioning'] +

    df_fe['prefarea']

)

print("Feature Created : luxury_score")

# -----------------------------------------------------
# Check New Features
# -----------------------------------------------------

new_features = [

    'total_rooms',

    'area_per_bedroom',

    'area_per_bathroom',

    'luxury_score'

]

print("\nNew Features")

print(new_features)

# -----------------------------------------------------
# Preview New Features
# -----------------------------------------------------

print("\nFeature Preview")

display(
    df_fe[new_features].head()
)

# -----------------------------------------------------
# Statistical Summary
# -----------------------------------------------------

print("\nFeature Statistics")

display(
    df_fe[new_features].describe()
)

# -----------------------------------------------------
# Correlation with Target
# -----------------------------------------------------

print("\nCorrelation with Price")

feature_corr = df_fe[
    new_features + ['price']
].corr()

display(
    feature_corr[['price']]
    .sort_values(
        by='price',
        ascending=False
    )
)

# -----------------------------------------------------
# Visualize Correlation
# -----------------------------------------------------

plt.figure(figsize=(8,5))

sns.heatmap(
    feature_corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title(
    "Feature Engineering Correlation Heatmap"
)

plt.show()

# -----------------------------------------------------
# Dataset Shape
# -----------------------------------------------------

print("\nDataset Shape After Feature Engineering")

print(df_fe.shape)

# -----------------------------------------------------
# Dataset Preview
# -----------------------------------------------------

print("\nDataset Preview")

display(df_fe.head())

# -----------------------------------------------------
# Summary
# -----------------------------------------------------

print("\nFEATURE ENGINEERING SUMMARY")

print("-"*50)

print(
    "Features Added :",
    len(new_features)
)

print(
    "Final Rows :",
    df_fe.shape[0]
)

print(
    "Final Columns :",
    df_fe.shape[1]
)

print("\nFeature Engineering Completed Successfully")
```

```python
# =====================================================
# STEP 7 : ENCODE REMAINING CATEGORICAL VARIABLES
# =====================================================

print("="*60)
print("ENCODE REMAINING CATEGORICAL VARIABLES")
print("="*60)

# -----------------------------------------------------
# Create Copy
# -----------------------------------------------------

df_model = df_fe.copy()

# -----------------------------------------------------
# Check Remaining Categorical Columns
# -----------------------------------------------------

print("\nCategorical Columns:")

categorical_cols = df_model.select_dtypes(
    include='object'
).columns

print(categorical_cols.tolist())

# -----------------------------------------------------
# Check Category Distribution
# -----------------------------------------------------

print("\nFurnishing Status Categories")

print(
    df_model['furnishingstatus']
    .value_counts()
)

# -----------------------------------------------------
# One-Hot Encoding
# -----------------------------------------------------

df_model = pd.get_dummies(
    df_model,
    columns=['furnishingstatus'],
    drop_first=True
)

# -----------------------------------------------------
# Verify New Columns
# -----------------------------------------------------

print("\nEncoded Columns Created")

new_cols = [

    col

    for col in df_model.columns

    if 'furnishingstatus' in col

]

print(new_cols)

# -----------------------------------------------------
# Dataset Information
# -----------------------------------------------------

print("\nDataset Information")

df_model.info()

# -----------------------------------------------------
# Check Remaining Object Columns
# -----------------------------------------------------

print("\nRemaining Object Columns")

remaining_cat = df_model.select_dtypes(
    include='object'
).columns

print(remaining_cat.tolist())

# -----------------------------------------------------
# Dataset Shape
# -----------------------------------------------------

print("\nDataset Shape")

print("Rows    :", df_model.shape[0])

print("Columns :", df_model.shape[1])

# -----------------------------------------------------
# Dataset Preview
# -----------------------------------------------------

print("\nDataset Preview")

display(df_model.head())

# -----------------------------------------------------
# Verify Encoded Columns
# -----------------------------------------------------

print("\nEncoded Column Sample")

display(

    df_model[
        [
            'furnishingstatus_semi-furnished',
            'furnishingstatus_unfurnished'
        ]
    ].head()

)

# -----------------------------------------------------
# Save Model Ready Dataset
# -----------------------------------------------------

df_model.to_csv(
    "Housing_Model_Ready.csv",
    index=False
)

print(
    "\nModel Ready Dataset Saved Successfully"
)

# -----------------------------------------------------
# Summary
# -----------------------------------------------------

print("\nENCODING SUMMARY")

print("-"*50)

print(
    "Total Features :",
    len(df_model.columns)-1
)

print(
    "Target Column : price"
)

print(
    "Remaining Categorical Columns :",
    len(remaining_cat)
)

print("\nStep 7 Completed Successfully")
```

```python
# =====================================================
# STEP 9 : TRAIN-TEST SPLIT & FEATURE SCALING
# =====================================================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib

print("="*60)
print("TRAIN-TEST SPLIT & FEATURE SCALING")
print("="*60)

# -----------------------------------------------------
# Separate Features and Target
# -----------------------------------------------------

X = df_model.drop('price', axis=1)

y = df_model['price']

print("\nFeatures Shape :", X.shape)
print("Target Shape   :", y.shape)

# -----------------------------------------------------
# Train-Test Split
# -----------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("\nTrain-Test Split Completed")

print("X_train Shape :", X_train.shape)
print("X_test Shape  :", X_test.shape)

print("y_train Shape :", y_train.shape)
print("y_test Shape  :", y_test.shape)

# -----------------------------------------------------
# Feature Scaling
# -----------------------------------------------------

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

# -----------------------------------------------------
# Convert Back To DataFrame
# -----------------------------------------------------

X_train_scaled = pd.DataFrame(
    X_train_scaled,
    columns=X.columns
)

X_test_scaled = pd.DataFrame(
    X_test_scaled,
    columns=X.columns
)

# -----------------------------------------------------
# Verify Scaling
# -----------------------------------------------------

print("\nScaled Training Dataset Preview")

display(X_train_scaled.head())

# -----------------------------------------------------
# Check Mean and Standard Deviation
# -----------------------------------------------------

print("\nFeature Means")

display(
    X_train_scaled.mean().round(2)
)

print("\nFeature Standard Deviations")

display(
    X_train_scaled.std().round(2)
)

# -----------------------------------------------------
# Save Scaler
# -----------------------------------------------------

joblib.dump(
    scaler,
    "scaler.pkl"
)

print("\nScaler Saved Successfully")

# -----------------------------------------------------
# Summary
# -----------------------------------------------------

print("\nTRAIN-TEST SPLIT SUMMARY")

print("-"*50)

print(
    "Training Records :",
    len(X_train)
)

print(
    "Testing Records :",
    len(X_test)
)

print(
    "Number of Features :",
    X.shape[1]
)

print("\nStep 9 Completed Successfully")
```

```python
# =====================================================
# STEP 10 : DATA MODELING
# =====================================================

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

print("="*60)
print("DATA MODELING")
print("="*60)

# -----------------------------------------------------
# 1. Linear Regression
# -----------------------------------------------------

lr_model = LinearRegression()

lr_model.fit(
    X_train_scaled,
    y_train
)

lr_pred = lr_model.predict(
    X_test_scaled
)

print("✓ Linear Regression Model Trained Successfully")

# -----------------------------------------------------
# 2. Decision Tree Regressor
# -----------------------------------------------------

dt_model = DecisionTreeRegressor(
    random_state=42
)

dt_model.fit(
    X_train_scaled,
    y_train
)

dt_pred = dt_model.predict(
    X_test_scaled
)

print("✓ Decision Tree Model Trained Successfully")

# -----------------------------------------------------
# 3. Random Forest Regressor
# -----------------------------------------------------

rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(
    X_train_scaled,
    y_train
)

rf_pred = rf_model.predict(
    X_test_scaled
)

print("✓ Random Forest Model Trained Successfully")

# -----------------------------------------------------
# Preview Predictions
# -----------------------------------------------------

comparison = pd.DataFrame({

    'Actual Price': y_test.values,

    'Linear Regression': lr_pred,

    'Decision Tree': dt_pred,

    'Random Forest': rf_pred

})

print("\nPrediction Comparison")

display(
    comparison.head()
)

print("\nAll Models Trained Successfully")
```

```python
# ===================================================
#                 Model Evaluation
# ===================================================

from sklearn.metrics import (mean_absolute_error, mean_squared_error, r2_score)

# Evaluation Function

def evaluate_model(model_name, y_true, y_pred):
    mae = mean_absolute_error(y_true,y_pred)
    mse = mean_squared_error(y_true,y_pred)
    rmse = np.sqrt(mse)

    r2 = r2_score(y_true, y_pred)
    return [model_name, mae, mse, rmse, r2]

results = []

results.append(
    evaluate_model(
        "Linear Regression",
        y_test,
        lr_pred
    )
)

results.append(
    evaluate_model(
        "Decision Tree",
        y_test,
        dt_pred
    )
)

results.append(
    evaluate_model(
        "Random Forest",
        y_test,
        rf_pred
    )
)

# Create Comparison Table

evaluation_df = pd.DataFrame(
    results,
    columns=[
        "Model",
        "MAE",
        "MSE",
        "RMSE",
        "R2 Score"
    ]
)

# Sort by Best R2

evaluation_df = evaluation_df.sort_values(
    by="R2 Score",
    ascending=False
)

print("\nMODEL PERFORMANCE COMPARISON\n")
print(evaluation_df)

# Best Model

best_model = evaluation_df.iloc[0]["Model"]
best_r2 = evaluation_df.iloc[0]["R2 Score"]

print("\nBest Model: ", best_model)
print("Best R2 Score: ",  round(best_r2,4))
```

```python
import tkinter as tk
from tkinter import ttk,messagebox
import pandas as pd
import joblib


# =====================================================
# LOAD MODEL FILES
# =====================================================

model = joblib.load("house_price_rf_model.pkl")
scaler = joblib.load("scaler.pkl")
feature_names = joblib.load("features.pkl")
#=====================================================
root =tk.Tk()
root.title("House price prediction dashboard")
root.geometry("1920x1080")
root.configure(bg="#EAF4FC")

#===================================================
# title
#===================================================
title = tk.Label(
    root,
    text="House price prediction system",
    font=("Arial",22,"bold"),
    bg="#1F4E79",
    fg="white",
    pady=15
)
title.pack(fill="x")
#===================================================
# form frame creation
frame = tk.Frame(root, bg="#EAF4FC")
frame.pack(pady=20)

#==================================================
#input fields
#===================================================
def add_label(row, text):
    tk.Label(
        frame,
        text=text,
        font=("Arial", 11),
        bg="#EAF4FC"
    ).grid(row=row, column=0, padx=10, pady=5, sticky="w")

# Area
add_label(0, "Area")
area_entry = tk.Entry(frame, width=20)
area_entry.grid(row=0, column=1)

# Bedrooms
add_label(1, "Bedrooms")
bedrooms_entry = tk.Entry(frame, width=20)
bedrooms_entry.grid(row=1, column=1)
# Bathrooms
add_label(2, "Bathrooms")
bathrooms_entry = tk.Entry(frame, width=20)
bathrooms_entry.grid(row=2, column=1)

# Stories
add_label(3, "Stories")
stories_entry = tk.Entry(frame, width=20)
stories_entry.grid(row=3, column=1)

# Parking
add_label(4, "Parking")
parking_entry = tk.Entry(frame, width=20)
parking_entry.grid(row=4, column=1)


# Dropdowns
def dropdown(row, text):
    add_label(row, text)

    box = ttk.Combobox(
        frame,
        values=["Yes", "No"],
        state="readonly",
        width=17
    )

    box.current(0)
    box.grid(row=row, column=1)

    return box

mainroad=dropdown(5, "Main Road")
guestroom = dropdown(6, "Guest Room")
basement = dropdown(7, "Basement")
hotwaterheating = dropdown(8, "Hot Water Heating")
airconditioning = dropdown(9, "Air Conditioning")
prefarea = dropdown(10, "Preferred Area")

# Furnishing
add_label(11, "Furnishing Status")
furnishing = ttk.Combobox(
    frame,
    values=[
        "furnished",
        "semi-furnished",
        "unfurnished"
    ],
    state="readonly",
    width=17
)

furnishing.current(0)
furnishing.grid(row=11, column=1)

# =====================================================
# RESULT LABEL
# =====================================================

result_label = tk.Label(
    root,
    text="",
    font=("Arial", 16, "bold"),
    bg="#EAF4FC",
    fg="green"
)

result_label.pack(pady=20)

# =====================================================
# PREDICTION FUNCTION
# =====================================================

def predict_price():

    try:

        area = float(area_entry.get())
        bedrooms = int(bedrooms_entry.get())
        bathrooms = int(bathrooms_entry.get())
        stories = int(stories_entry.get())
        parking = int(parking_entry.get())

        mainroad_val = 1 if mainroad.get() == "Yes" else 0
        guestroom_val = 1 if guestroom.get() == "Yes" else 0
        basement_val = 1 if basement.get() == "Yes" else 0
        hotwater_val = 1 if hotwaterheating.get() == "Yes" else 0
        air_val = 1 if airconditioning.get() == "Yes" else 0
        pref_val = 1 if prefarea.get() == "Yes" else 0

        total_rooms = bedrooms + bathrooms

        area_per_bedroom = area / bedrooms

        area_per_bathroom = area / bathrooms

        luxury_score = (
            guestroom_val +
            basement_val +
            hotwater_val +
            air_val +
            pref_val
        )

        semi_furnished = 0
        unfurnished = 0

        if furnishing.get() == "semi-furnished":
            semi_furnished = 1

        elif furnishing.get() == "unfurnished":
            unfurnished = 1

        input_df = pd.DataFrame({

            'area':[area],
            'bedrooms':[bedrooms],
            'bathrooms':[bathrooms],
            'stories':[stories],
            'mainroad':[mainroad_val],
            'guestroom':[guestroom_val],
            'basement':[basement_val],
            'hotwaterheating':[hotwater_val],
            'airconditioning':[air_val],
            'parking':[parking],
            'prefarea':[pref_val],
            'total_rooms':[total_rooms],
            'area_per_bedroom':[area_per_bedroom],
            'area_per_bathroom':[area_per_bathroom],
            'luxury_score':[luxury_score],
            'furnishingstatus_semi-furnished':[semi_furnished],
            'furnishingstatus_unfurnished':[unfurnished]

        })

        input_df = input_df.reindex(
            columns=feature_names,
            fill_value=0
        )

        scaled_input = scaler.transform(input_df)

        prediction = model.predict(scaled_input)

        result_label.config(
            text=f"Predicted House Price: ₹ {prediction[0]:,.2f}"
        )

    except Exception as e:

        messagebox.showerror(
            "Error",
            str(e)
        )

#=========================================================================
#button
#=========================================================================

predict_btn=tk.Button(
    root,
    text="Predict house price",
    font=("Arial",14,"bold"),
    bg="#28A745",
    fg="white",
    command=predict_price
)
predict_btn.pack(pady=20)

root.mainloop()
```
